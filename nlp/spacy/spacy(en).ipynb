```
import spacy
nlp = spacy.load("en_core_web_sm")
```

```
# Текст для анализа
text = """London is the capital and largest city of England and the United Kingdom. 
The city stands on the River Thames in the south-east of England, at the head of its 50-mile (80 km) estuary leading to the North Sea."""

# Парсинг текста с помощью spaCy. Выводим текст токена, начальная форма, часть речи, является ли стоп-словом
doc = nlp(text)
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.is_stop)
```

```
# Построение дерева зависимостей
from spacy import displacy
displacy.render(doc[:11], style='dep', jupyter=True)
```

```
# Распознание именованных сущностей 
text2 = """ Hello! How are you?
Hi! I am fine, thanks! And how are you?
Fine! I am Jack. What is your name?
I am William, nice to meet you.
Nice to meet you too.
Are you from Barcelona?
No, I am from London. And you?
I am from USA. Nice to meet you in Spain!"""

doc2 = nlp(text2)
# for ent in doc2.ents:
#     print(ent.text, ent.label_)
displacy.render(doc2, style='ent', jupyter=True)
```
