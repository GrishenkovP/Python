```
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier, CatBoostRegressor, CatBoost, Pool
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
df = pd.read_csv("/content/drive/MyDrive/db/iris_frame2.csv", sep=",")
df.head()
```

CatBoostRegressor

```
# initialize data
train_data = df.loc[:, df.columns != 'target'].values
train_labels = df.loc[:, 'target'].values


# scaler = MinMaxScaler((-1, 1))
# train_data = scaler.fit_transform(train_data)

test_data = catboost_pool = Pool(train_data, 
                                 train_labels)

# model = CatBoostClassifier() - дает accuracy_score равное 1
model = CatBoostClassifier(iterations=2,
                           depth=1,
                           learning_rate=1,
                           loss_function='MultiClass',
                           verbose=True)


# train the model
model.fit(train_data, train_labels)
# make the prediction using the resulting model
preds_class = model.predict(test_data)

print(accuracy_score(test_data.get_label(), preds_class))
```

CatBoostRegressor

```
# initialize data
data = df.loc[:, df.columns != 'target'].values
labels = df.loc[:, 'target'].values


# scaler = MinMaxScaler((-1, 1))
# data = scaler.fit_transform(data)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.14, random_state=7)

# initialize Pool
train_pool = Pool(train_data, 
                  train_labels)
test_pool = Pool(test_data) 

# specify the training parameters 
# model = CatBoostRegressor() дает accuracy_score равное 0,9
model = CatBoostRegressor(iterations=3, 
                          depth=1, 
                          learning_rate=1, 
                          loss_function='RMSE')
#train the model
model.fit(train_pool)
# make the prediction using the resulting model
preds = model.predict(test_pool)
preds = [round(pred) for pred in preds]

print(accuracy_score(test_labels,preds))
```

CatBoost

```
# initialize data
data = df.loc[:, df.columns != 'target'].values
labels = df.loc[:, 'target'].values


# scaler = MinMaxScaler((-1, 1))
# data = scaler.fit_transform(data)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.14, random_state=7)

# initialize Pool
train_pool = Pool(train_data, 
                  train_labels)
test_pool = Pool(test_data) 

# specify training parameters via map

param = {'iterations':3}
# model = CatBoost() дает accuracy_score равное 0,9 
model = CatBoost(param)
#train the model
model.fit(train_pool) 

# make the prediction using the resulting model
preds_class = model.predict(test_pool)

preds_class = [round(pred_class) for pred_class in preds_class]

print(accuracy_score(test_labels,preds_class))
```
